```python
import numpy as np
import networkx as nx
import pickle
```

```python
links = np.genfromtxt('facebook.txt', delimiter=' ', dtype=int)
g = nx.Graph()
g.add_edges_from(links)
net = nx.adjacency_matrix(g).toarray().astype(int)
for i in range(net.shape[0]):
    net[i, i] = 1
```

```python
# tst.test_stats(g)
```

```python
alpha0 = 2.0
sigma0 = .98
c0 = 100.
lamb0 = 1000.

model = spd.NetworkFull(net, links, alpha=alpha0, sigma=sigma0, c=c0, lamb=lamb0)
```

```python
with open('prefit.pkl', 'rb') as fn:
    presampled = pickle.load(fn)
model.K = presampled.num_cliques
Z = np.zeros(model.num_cliques, model.num_nodes)
for i, cl in enumerate(presampled.cliquetonodes):
    for node in cl:
        Z[i, node] = 1
model.Z = Z
```

```python
save_every = 1
write_every = 1

niter = 25000

# model.clique_init()

model.sample(
    num_iters= niter,
    save_every = save_every,  
    write_every = write_every,
    verbose = True,
    sample_hypers = False,
    optim_hypers = True,
    hyper_start = 0,
    num_sm = 50,
    num_hyper = 5,
    do_gibbs = False,
    write_Z=True,
    filename='facebook.pkl')   
```
